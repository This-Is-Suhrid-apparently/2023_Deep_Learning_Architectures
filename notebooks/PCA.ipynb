```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

## The data is 30-dimensional. We have to reduce it to 2 dimensions using PCA

```python
#Creating Dataframe
df = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
```

```python
df.head(5)
```

```python
#While implementing PCA, always apply standard Scaling first
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)

#Can also use MinMax scaler
# from sklearn.preprocessing import MinMaxScaler
# scaler = MinMaxScaler()
# scaler.fit(df)
```

```python
#To finish the scaling, we need to transform the scaler
scaled_data = scaler.transform(df)
scaled_data #Will have data with standard deviation = 1 and mean = 0
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2) #Selecting the number of components to be decomposed to
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
scaled_data #Shows PCA decomposed data
```

```python
x_pca
```

```python
#Plotting the data in the Principal Components
plt.figure(figsize = (8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c=cancer['target'])
plt.xlabel('First Priciple Component')
plt.ylabel('Second Principle Component')
```

### It is clear that even after decomposing the 30-dimensional data into 2 dimensions, we still get data that is separated into 2 categories.
